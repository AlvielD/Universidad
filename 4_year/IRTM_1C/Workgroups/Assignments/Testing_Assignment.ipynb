```python
from typing import List, Union
import pytest
import ipytest

ipytest.autoconfig()
```

```python
def get_unique_elements(
    lst: List[Union[str, int]], n: int = 1
) -> List[Union[str, int]]:
    """Given a list of elements returns those that repeat at least n times. The
    output list should contain all unique elements and they should be returned
    in the same order as they first appear in the input list.

    Args:
        lst: Input list
        n (optional): Minimum number of times an element should be repeated to 
            be returned. Defaults to 1.

    Returns:
        List of unique items 
    """
    # TODO
    
    i = 0
    r_lst = []
    nextElem = False

    # For each element in the list
    while i in range(len(lst)):
        
        # By default, we check the element
        nextElem = False

        # Search in the result if the element exist
        for k in range(len(r_lst)):
            if lst[i] == r_lst[k]:
                nextElem = True

        # If the element is not in the result list, then we proceed
        if nextElem == False:

            j = i+1     # Start counting on the next element
            times = 1   # Count the element

            # We check if there is another identical element 
            while j in range(len(lst)) and times < n:

                # Count the element
                if lst[i] == lst[j]:
                    times += 1
                
                # If it repeats at least n times, we append it to the list
                if times == n:
                    r_lst.append(lst[i])

                j += 1

        i += 1
        
    return r_lst
```

```python
%%run_pytest[clean]

#from A0 import get_unique_elements
@pytest.fixture


def lst():
    return [1, 1, 1, 1, 2, 2, 2, 3, 3, 3, "test1", "test2", "test2"]


def test_get_unique_elements_empty():
    assert get_unique_elements([]) == []


def test_get_unique_elements_default(lst):
    assert get_unique_elements(lst) == [1, 2, 3, "test1", "test2"]


def test_get_unique_elements_n_2(lst):
    assert get_unique_elements(lst, 2) == [1, 2, 3, "test2"]


def test_get_unique_elements_n_3(lst):
    assert get_unique_elements(lst, 3) == [1, 2, 3]


def test_get_unique_elements_n_10(lst):
    assert get_unique_elements(lst, 10) == []
```
